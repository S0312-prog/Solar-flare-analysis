```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
import xspec
from xspec import *
```

```python
Plot.device = '/xw'    
Plot.xAxis = 'keV'
```

```python
xsm="paper/ch2_xsm_20240912_1.pha"
AllData.clear()
AllModels.clear()
spec_xsm=xspec.Spectrum(xsm)
spec_xsm.ignore("**-2.2 10.0-**")
model_xsm=xspec.Model('vvapec')
model_xsm(17).frozen=False
model_xsm(27).frozen=False
model_xsm(29).frozen=False
model_xsm(21).frozen=False
model_xsm(19).frozen=False
xspec.Fit.perform()
Plot("ldata delchi")
spect_xsm= np.array(xspec.Plot.y(1))
```

```python
solexs="paper/Al1_20240912_1_corrected.pi"
AllData.clear()
AllModels.clear()
spec_solexs=xspec.Spectrum(solexs)
spec_solexs.ignore("**-2.2 10.0-**")
model_xsm=xspec.Model('vvapec')
model_xsm(17).frozen=False
model_xsm(27).frozen=False
model_xsm(29).frozen=False
model_xsm(21).frozen=False
model_xsm(19).frozen=False
xspec.Fit.perform()
Plot("ldata delchi")
spect_solexs= np.array(xspec.Plot.y(1))
```

```python
plt.plot(spect_xsm,spect_solexs)
plt.show()
```
